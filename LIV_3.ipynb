## 2- Explication détaillée des étapes :

- Nous pouvons généralement diviser la chapne de transmssion en 3 bornes majeures, elle sont :

### **1. Borne d'émission :**

**Entrée (haut parleur) :** L'outil utilisé pour enregistrer le message, et cela sous forme analogique, sa fréquence d'echantillonage est de 44100Hz

**Ordinateur de la base :**

  - CAN : Une conversion de l'analogique au binaire, il est obligatoire car le codage néccessite un signal de type binaire elle se compose de deux sous-étapes:

    - **Quantification :** Chaque échantillon du micro est une valeur comprise dans l'intervalle ]0;1[. Et est situé dans une liste de petits intervalles chaqun représenté par des bits, la taille de ces intervalles est décidée par la taille de la tramme choisie.

  - Codage du message bianire : Utilisation de l'algorithme Manchester

**Modulation :** Une méthode d'augmentation artificielle de la fréquence du message de la fréquence définie (32000Hz) vers de ultra hautes (fréquences du canal d'émission)

### **2. Borne de transmission :**

**Antenne d'émission :** Le dispositif utilisé pour émettre le message (maintenant à haute fréquence) dans le canal à haute fréquence dans lequel le message sera transmis, la taille de l'an

**Canal de transmission :** Une collection de hautes fréquences, il comporte un grand nombre de fréquences la pluspart sont indésirables, une d'entre eux est le fréquence cible (la fréquence du message)

**Filtrage :** Le canal de transmission est filtré (avec le filtre passe haut proposé en livrable 3), éliminant les fréquences indésirables et gardant la fréquence cible

**Antenne de réception :** En pratique, le filtrage prend place dans l'antenne, qui est le dispositif de réception du message en haute fréquence

- L'existence même de la modulation (et donc l'émission dans un canal de transmission) revient à le taille de l'antenne qui devrait être gardée raisonnable. Car en effet la taille requise de l'antenne grandit inversement avec la fréquence d'émission, des communication à longues distances néccessiterait des antennes impossiblement grandes

### **3. Borne de réception :**

**Démodualtion :** Le procédé inverse de la modulation, et elle se résume en la baisse de la fréquence du message reçu vers la fréquence choisie de base (32000Hz)

**Ordinateur de l'agence (ordi d'écoute):**

  - CAN : Une conversion de l'analogique reçu de l'antenne au numérique requis par l'algorythme de décodage

  - Décodage du message binaire : (possiblement manchester)

  - CNA : Une conversion de la sortie numéruque de l'algorythme de codage vers l'analogique requis pour le lecture du message à travers le haut-parleur.

**Sortie :** La sortie choisie est un haut-parleur de l'agence, qui prendra naturellement un signal analogique (ce qui justifie la CNA précédente)

## Expemple de la chaîne sur un seul caractère :

- Exemple sur un caractère :
- Prenons une valeur arbitraire de 0.665 comme entrée, et convertissons la en binaire :
bin(0.665) = 680 = 0000001010101000
- On effectue un codage Manchester qui nous donne 0000000000111000
en le décode à l'arrivée pour obtenir 0000001010101000
on effectue une conversion binaire analogique
0000001010101000 = 679.8 = 680
analog(680) = 0.665
- Et on obtient la valeur originale
