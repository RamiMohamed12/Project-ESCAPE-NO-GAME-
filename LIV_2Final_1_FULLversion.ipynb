## 1. Proposer un schéma de filtre, et le justifier :

Notre but avant tout est de réussir à filtrer un message d'une grande bande de fréquences possibles. considérons que le fréquence cible (celle du message de l'agent) sera de $ 32000Hz $. Nous aurons donc besoin d'un circuit électrique pouvant atténuer toute distortion à fréquence plus basse que la fréquence cible.

Cet objectif prouve d'être un contexte parfait pour l'utilisation d'un filtre **Passe-haut**.


Dans un circuit RLC, la résistance (R), l'inductance (L) et la capacité (C) interagissent pour créer des réponses différentes en fonction de la fréquence du signal.

Dans un filtre passe-haut, on veut permettre le passage des fréquences élevées tout en atténuant les basses fréquences. Le circuit RLC est adapté pour cela car à certaines fréquences, la réactance de l'inductance (XL) ou de la capacité (XC) peut dominer le comportement du circuit.

À une fréquence plus élevée, la réactance de la capacité (XC = 1 / (2πfC)) diminue, permettant ainsi le passage du signal à haute fréquence tandis que la réactance de l'inductance (XL = 2πfL) peut devenir élevée, limitant le passage des basses fréquences.

En contrôlant les valeurs des composants (R, L et C) dans le circuit RLC, on peut ajuster la fréquence de coupure où le filtre commence à atténuer les fréquences plus basses tout en laissant passer les fréquences plus élevées.

La façon dont les composants sont interconnectés modifie les propriétés électriques du circuit, affectant ainsi la réponse en fréquence. Par exemple, dans un filtre passe-haut du premier ordre, si la résistance est placée en série avec l'inductance ou la capacité, cela peut influencer la pente de coupure du filtre et les caractéristiques d'atténuation des basses fréquences.

En augmentant l'ordre du filtre (en ajoutant plus de composants ou en les interconnectant différemment), on peut obtenir des caractéristiques de filtrage plus prononcées et une atténuation plus forte des basses fréquences tout en laissant passer les hautes fréquences.

## 2- Étude du circuit :
### **2.a Donner l'impédance équivalente du circuit :**
On sait que : $ Z_R=R   /   Z_c= \frac{1}{jCω}  /   Z_L= jLω $
Ou :

- **L** est l'inductance en Henrys (H).

- **Z** est l'impédance en ohms (Ω).

- **ω** est la pulsation en radians par seconde (rad/s).

- **C** est la capacité en farads (F).

Pour donner l'impédance équivalente on additionne l'impédance de chaque composant :

$ Z_{éq}= Z_R + Z_C + Z_L$

$ Z_{éq}= R + \frac{1}{jCw} + jLω $

$ Z_{éq}= R + \frac{-j}{Cw} + jLω $

$\boxed{ Z_{éq}= R + j(\frac{-1}{Cw} +Lω) }$

### **2.b étudier les variations de son module en fonction de ω (pulsation) :**

Le module de l'impédance $∣Z_{éq}∣ $ est donné par :

$ \boxed{∣Z_{éq}∣ = \sqrt{[R^2 + (-\frac{1}{Cω} + Lω)^2 ]} }$

Pour étudier les variations, examinons différents cas :

**Basse fréquence ($ ω \rightarrow 0 $) :**

Lorsque la pulsation $ ω $ tend vers zéro, $ ω_L $ devient négligeable, et la formule devient :

$ ∣Z_{éq}∣ = \sqrt{[R^2 + (\frac{1}{Cω})^2]} $ 

>Donc $∣Z_{éq}∣ ➜ +\infty $

- L'impédance converge vers R, et son module de devient principalement déterminé par la valeur de la résistance à basse fréquence.

**Haute fréquence ($ ω \rightarrow \infty $) :**

À mesure que la pulsation $ ω $ augmente, le terme $ ω_L $ devient dominant, et $ -\frac{1}{Cω} $ devient négligeable, Au final on obtient :

$ |Z_{éq}| = \sqrt{(R^2 +(Lω)^2 )} $

>$ |Z_{éq}| ➜ +\infty $

- À haute fréquence, l'impédance est principalement déterminée par la réactance inductive ($ ω_L $).

**Fréquence de résonance ($ ω = ω_0 $) :**

- À la fréquence de résonance, la réactance inductive est égale à la réactance capacitive, annulant le terme 
$ ωL−\frac{1}{ωC} $ dans la formule. Ainsi, le module de l'impédance est égal à :

>$ |Z_{éq}|= R $

- Comme vu précédemment, à une fréquence de résonance ($ w = w_0 $):

$ ω_0L = \frac{1}{w_0C} $

$ ω_0^2 = \frac{1}{LC} $

$ ω_0 = \sqrt{\frac{1}{LC}} = \frac{1}{\sqrt{LC}} $

- On sait que :

#### $ \boxed{ω_0 = 2{\pi f_0}} $

- Donc:

$ \frac{1}{\sqrt{LC}} = 2\pi f_0 $

- Enfin la fréquence de résonance du circuit est : 

$ \boxed{f_0 =\frac{1}{2 \pi \sqrt{LC}}} $


### **5.c Déduire son coefficient de qualité Q :**

- La formule générale du coefficient de qualité est donnée par :

$ \boxed{Q = \frac{1}{R}{\sqrt{\frac{L}{C}}}} $  

- Ou bien :

$ \boxed{Q = \frac{1}{RCω_0}} $

## 3. Dimentionnement des composants :

## 4. Établir l'équation différentielle de l’évolution de la tension $ U_C $ aux bornes du condensateur, et trouver la solution numérique :
### **4.a L'équation différentielle :**
- Nous savons que :
$ U_R = Ri / U_L = L \frac{di}{dt} / i = C\frac{dU_C}{dt} $
- En appliquant la loi des mailles sur le circuit RLC on obtient :

$ E = U_R + U_C + U_L $

- En remplacant :

$ Ri + U_C + L\frac{di}{dt} = E $

$ RC\frac{dU_C}{dt}+L\frac{di}{dt}+U_C=E $

$ RC\frac{dU_C}{dt}+L\frac{d^2U_C}{dt^2}+U_C=E $

$ LC\frac{d^2U_C}{dt^2}+RC\frac{dU_C}{dt}+U_C=E $

$ \frac{d^2U_C}{dt^2}+\frac{R}{L}\frac{dU_C}{dt}+\frac{U_C}{LC}=\frac{E}{LC} $

- En admettant que :

$ w_0=\frac{1}{LC} / \frac{w_0}{Q} = \frac{R}{L} $

$ \boxed{\frac{d^2U_C}{dt^2}+\frac{w_0}{Q}\frac{dU_C}{dt}+w_0^2U_C=Ew_0^2} $

### **4.b Solution de l'équation :**

On calcule d'abord le Δ :

$Δ =b^2 -4(ac) w_0^2 $ 

$Δ = (\frac{w_0}{Q})^2 -4(w_0^2)$

On remplace $w_0 et Q$ par leurs valeurs ducoup on obtient :

$Δ = (\frac{64000\pi}{ \frac{1}{\sqrt{2}}})^2 -4((64000\pi)^2)$

>$Δ = -8*10^10 < 0$

On calcule les 2 solutions $r_1$ et $r_2$ :

$ r_1 = \frac{-b-j\sqrt{Δ}}{2a}$ 

$r_2 = \frac{-b+j\sqrt{Δ}}{2a} $

$ r_1 = \frac{-\frac{64000\pi}{\frac{1}{\sqrt{2}}}-j\sqrt{-8*10^10}}{2}$

$ r_2 = \frac{-\frac{64000\pi}{\frac{1}{\sqrt{2}}}+j\sqrt{-8*10^10}}{2}$

>$\boxed{ r_1 =-142172.254+142172.254j}$

>$\boxed{ r_2 =-142172.254-142172.254j}$

Donc voici la solution homogène : 

$\boxed{Y_H = exp(-142172.254t)(C_1\cos(142172.254t)+C_2\sin(-142172.254t))}$

#### Trouver la solution particulière :

$Ew_0^2$ est un polynome de degrés 0 çela signifie que:

$Y_P = A$

Donc la premiére dérivée et la deuxiéme sont nulles.

On remplace dans l'équation différentielle : 

$Aw_0^2 = Ew_0^2$

>$A=E$

Donc voici la solution particulière:

$\boxed{Y_H = exp(-142172.254t)(C_1\cos(142172.254t)+C_2\sin(-142172.254t)) + E}$

On fixe notre E à $5v$

>$\boxed{Y_H = exp(-142172.254t)(C_1\cos(142172.254t)+C_2\sin(-142172.254t)) + 5}$

## 5. Fonction de transfert, gain, et diagramme de Bode :
### **5.a Établir la fonction de transfert :**

- En appliquant le loi du pont diviseur de tension, on obtient :

$ V_s = \frac{Z_L}{Z_L + Z_C + Z_R}V_e $

$ V_s = \frac{jLω}{jLω +R + \frac{1}{jCω}}V_e $

$ V_s = \frac{jLω}{jLω +  \frac{1+RjCω}{jCω}}V_e $

$ V_s = \frac{jLω}{ \frac{-LCω^2+1+RjCω}{jCω}}V_e $

$ V_s = \frac{-LCω^2}{-LCω^2+1+RjCω}V_e $

On sait que :

$ω_0 = \frac{1}{\sqrt{LC}}$

>$LC= \frac{1}{ω_0^2}$

$Q=\frac{1}{RCω_0}$

$RC = \frac{1}{Qω_0}$

On remplaçant on obtient :

$ V_s = \frac{(-\frac{ω^2}{ω_0^2})}{-\frac{ω^2}{ω_0^2} +1 + j\frac{ω}{Qω_0}} V_e $

La fonction de transfert est donnée sous :

$ T(w) = V_s / V_e $

Ainsi :

$ \boxed{T(w) =\frac{-\frac{ω^2}{ω_0^2}}{-\frac{ω^2}{ω_0^2} +1 + j(\frac{ω}{Qω_0})}} $

Pour étudier le comportement de $ T(w) $ en fonction de la fréquence on prend les cas où f tend vers l'infini et vers 0.

### **5.b Étudier son comportement en fonction de la fréquence:**

On sait que $ w = 2\pi f$ donc il suffit de prendre les cas où w tend vers l'infini ou vers 0.

**Haute fréquence ($ ω \rightarrow \infty $) :**

Quand $ ω \rightarrow \infty $ : $ T(w) $ devient :

>$\boxed{T(w)= \frac{\frac{-(w_0)^2}{w^2}}{\frac{-(w_0)^2}{w^2}} = 1}$

**Basse fréquence ($ ω \rightarrow 0 $) :**

Quand $ ω \rightarrow 0 $ : $ T(w) $ devient :

>$\boxed{T(w) =\frac{0}{1} = 0}$

- Cela indique une atténuation des basses fréquences, conforme aux propriétés d'un filtre passe-haut.

**Fréquence de résonance ($ ω = ω_0 $) :**

Quand $ ω = ω_0 $ : $ T(w) $ devient :

>$\boxed{T(w) =\frac{-1}{\frac{j}{Q}}=\frac{-Q}{j} =jQ}$

- Cela signifie que la fonction de transfert tend vers une valeur maximale, représentant une réponse significative à la fréquence de résonance. 

### **5.c Déduire le gain (dB) (en décibels) :**

La formule du Gain_dB est donnée sous l'équation suivante :

$ Gain = 20log(|T(w)|) $

Pour déduire le Gain nous devons établir la fonction du module de la fonction de transfert T(w)

$ (|T(w)|) = √(-(ω/ω_0)^2)^2 /√((-(ω/ω_0)^2 + 1)^2 +(ω/Qω_0)^2) $

$ (|T(w)|) =(ω/ω_0)^2/√((-(ω/ω_0)+1)^2 +(ω/Qω_0)^2) $

Maintenant on remplace :

$Gain = 20log((ω/ω_0)^2/√((-(ω/ω_0)^2 +1)^2 +(ω/Qω_0)^2))$

$Gain = 20log(ω/ω_0)^2 - 20log(√((-(ω/ω_0)^2 +1)^2 +(ω/Qω_0)^2))$

>$Gain = 40log(ω/ω_0) -10log((-(ω/ω_0)^2 +1)^2 +(ω/Qω_0)^2)$

Pour tracer le diagramme de Bode qui représente le Gain on doit étudier les 3 cas suivants :

**a. $ w>>ω_0 $:**

Le 1 devinet négligeable, on prend $(ω/ω_0)^2$ comme facteur en commun on obtient :

$Gain = 40log(ω/ω_0) -10log((ω/ω_0)^2[(ω/ω_0)^2  +(1/Q))]$

1/Q devinet négligable,donc:

$Gain = 40log(ω/ω_0) -10log((ω/ω_0)^4)$

$Gain = 40log(ω/ω_0) -40log(ω/ω_0) $

>$Gain =0 db$ 

**b. $ w=wω_0 $:**

$ Gain = 40log(1) -10log((-1 +1)^2 +(1/Q^2)) $

$ Gain = -10log(1) + 10log(Q^2) $

>$ Gain = 20log(Q) = -3 db $

**c. $ w<<ω_0 $:**

$(ω/Qω_0)^2$ et $-(ω/ω_0)^2$ deviennent négligeable face au 1, donc :

$Gain = 40log(ω/ω_0) -10log(1)

>$Gain = 40log(ω/ω_0)$




### **5.d Déduire l'argument (phase) :**

pour déduire la phase nous devons trouver l'équation de l'arguments de la fonction de transfert :

$Arg(T) =Arg(-(ω/ω_0)^2/-(ω/ω_0)^2 +1 + j(ω/Qω_0))$

$Arg(T) = Arg(-(ω/ω_0)^2) - Arg(-(ω/ω_0)^2 +1 + j(ω/Qω_0))$

>On sait que $Arg(x + iy) = Arctan(iy/x)$

$Arg(-(ω/ω_0)^2)$ La partie imaginaire est égal à 0 et la patie réel est négatif cela veut dire que $Arg(-(ω/ω_0)^2) = π $

$Arg(T) = π$
- Arg(-(ω/ω_0)^2 +1 + j(ω/Qω_0))$

$Arg(T) = π -Arctan((ω/Qω_0)/(1-(ω/ω_0)^2))$

**a- $w>>ω_0$ :**

Le 1 devient négligeable donc $Arctan((ω/Qω_0)/1-(ω/ω_0)^2)$ = $Arctan((ω/Qω_0)/-(ω/ω_0)^2)$

$Arg(T) = π -Arctan(-ω_0/ω*Q)$

>+$Arg(T) = π - π = 0$

**b- $w=ω_0$ :**

$1-(ω/ω_0)^2$ devient 0 donc la partie réel est égal à 0 et la partie imaginaire $(1/Q)$ est positif donc :

$Arctan((ω/Qω_0)/1-(ω/ω_0)^2) = π/2 $

>$Arg(T) = π -π/2 = π/2$

**c- $w<<ω_0$ :**

Quand w<<ω0 çela veut dire que $(ω/ω_0) ➜ 0$ donc la fonction devient :

$Arg(T) = π -Arctan(0/1)$
>$Arg(T) = π -0= π $


**Les deux diagrammes do bode :**

```python
import numpy as np
import matplotlib.pyplot as plt

def high_pass_filter_bode_plot(omega, R, L, C, omega_0, Q):
    omega_02 = omega**2
    numerator = -omega_02 / (omega_0**2)
    denominator = (-omega_02 / (omega_0**2)) + 1 + 1j * (omega / (Q * omega_0))

    H = numerator / denominator
    magnitude = np.abs(H)
    phase = np.angle(H)

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

    ax1.semilogx(omega, 20 * np.log10(magnitude), label='Gain (dB)')
    ax1.set_ylabel('Gain (dB)')
    ax1.grid(True)
    ax1.legend()

    ax2.semilogx(omega, phase * 180 / np.pi, color='red', label='Phase (degrees)')
    ax2.set_xlabel('Fréquence(Hz)')
    ax2.set_ylabel('Phase (degrees)')
    ax2.legend()

    plt.title("Graph De Le Gain et De La Phase")
    plt.tight_layout()
    plt.show()

R = 28000 
L = 0.1  
C = 2.47 * 10**-10  
omega_0 = 1 / np.sqrt(L * C)
Q = 1 / np.sqrt(2)

omega = np.logspace(1, 10, num=100000)
high_pass_filter_bode_plot(omega, R, L, C, omega_0, Q)
```
