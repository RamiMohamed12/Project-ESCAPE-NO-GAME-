### 1- Proposer un schéma de filtre  permettant de récupérer le signal non audible à partir du signal reçu
`Partie de Racim`

## 2- Etude du circuit :
### `a- Donner l'impédance équivalente du circuit:`
On sait que  $Z_R=R$    et que     $Zc= 1/jcω $   et que $Z_L= jLω$

L est l'inductance en Henrys (H),

Z est l'impédance en ohms (Ω),

ω est la pulsation en radians par seconde (rad/s),

C est la capacité en farads (F).

Pour donner l'impédance équivalente on additionne l'impédance de chaque composant :

$Zeq= Z_R + Zc + Z_L$

$Zeq= R + 1/jcω  + jLω$ 

$ Zeq= R - j/cω + jLω$

>$ Zeq= R + j(-1/cω +Lω) $

#### `b- étudier les variations de son module en fonction de ω (pulsation):`

Le module de l'impédance ($∣Zeq∣$) est donné par :

>($∣Zeq∣$) =  $√(R^2 + (-1/cω +Lω)^2 )$  

Pour étudier les variations, examinons différents cas :

####  `1- Basse fréquence (ω  ➜  0)`

Lorsque la pulsation ω tend vers zéro,ωL devient négligeable, et la formule devient :

>($∣Zeq∣$) =  $√(R^2 + (1/cω)^2 )$            

Cela converge vers la résistance R, et le module de l'impédance devient principalement déterminé par la résistance à basse fréquence.

#### `2-Haute fréquence (ω  ➜  ∞ ) :`

À mesure que la pulsation ω augmente, le terme ωL devient dominant, et $-1/cω$ devient négligeable, Au final on obtient :

>($|Zeq|$) = $√(R^2 +Lω^2 )$

À haute fréquence, l'impédance est principalement déterminée par la réactance inductive (ωL).

#### `Fréquence de résonance ( ω = ω0 ) :`

À la fréquence de résonance, la réactance inductive est égale à la réactance capacitive, annulant le terme 
$ωL− 1/ωC$ dans la formule. Ainsi, le module de l'impédance est égal à :

>($|Zeq|$)= $R$


#### `c- Déduire la fréquence de résonance`

Comme déja vue précédemment quand  $ωL− 1/ωC$ s'anullent donc ça veut dire que $ωL = 1/ωC$ et dans çe cas ω = ω0 ça devient :

$ω_0L = 1/ω0C$

$ω_0^2 = 1/LC$

$ω_0 = √(1/LC)$

>$ω_0 = 1/√LC$

On sait que :

#### $ω_0 = 2πf_0 $

Donc:

$1/√LC = 2πf_0 $

Enfin la fréquence de résonance du circuit est : 

>$f_0 =1/ 2π√LC $

On considere les valeurs des composants 

$R=200 Ω$
   
$L=0.1H$

$C=10µF$

Pour déterminer $f_0$ on remplace et on obtient :

$f_0 =1/ 2π√0.1*(10*10^-6) $

>$f_0 = 159.1 HZ $

#### `d- Déduire son coefficient de qualité Q : `

La formule générale du coefficient de qualité est donnée par :

>$Q=1/R√(L/C)$  

Ou bien :

>$Q=1/RCω_0$

On sait que $f_0 = 159.1 HZ $, donc on peut trouvé la valuer de $ω_0$ :

$ω_0 = 2πf_0 $

$ω_0 = 2π*159.1$

$ω_0= 1000  rd/s$

Donc:

$Q=1/200*(10^-5)*1000$

>$Q=0.5$


## 5 - Donner la fonction de transfert T(w) du filtre proposé:

On applique la loi du pont diviseur de tension on obtient :

$V_S = (Z_L / Z_L + Z_C + Z_R)V_e$

$V_S = (jLω/jLω +R + 1/jCω)V_e$

$V_S = (jLω/jLω + (1+RjCω/jCω))V_e$

$V_S = (jLω/(-LCω^2+1+RjCω)/jCω)V_e$

$V_S = (-LCω^2/-LCω^2+1+RjCω)V_e$

On sait que :

$ω_0 = 1/√LC$

>$LC= 1/ω_0^2$

$Q=1/RCω_0$

$RC = 1/Qω_0$

On remplaçant on obtient:

>$V_S = (-(ω/ω_0)^2/-(ω/ω_0)^2 +1 + j(ω/Qω_0)) V_e$

La fonction de transfert est donnée sous :

>$T(w) = V_s / V_e $

Ainsi :

>$T(w) =-(ω/ω_0)^2/-(ω/ω_0)^2 +1 + j(ω/Qω_0)$

#### `Déduire le Gain_dB (en décibels):`

La formule du Gain_dB est donnée sous l'équation suivante :

>$Gain = 20log(|T(w)|)$

Pour déduire le Gain nous devons établir la fonction du module de la fonction de transfert T(w)

$(|T(w)|) = √(-(ω/ω_0)^2)^2 /√((-(ω/ω_0)^2 + 1)^2 +(ω/Qω_0)^2)$

$(|T(w)|) =(ω/ω_0)^2/√((-(ω/ω_0)+1)^2 +(ω/Qω_0)^2)$

Maintenant on remplace :

$Gain = 20log((ω/ω_0)^2/√((-(ω/ω_0)^2 +1)^2 +(ω/Qω_0)^2))$

$Gain = 20log(ω/ω_0)^2 - 20log(√((-(ω/ω_0)^2 +1)^2 +(ω/Qω_0)^2))$

>$Gain = 40log(ω/ω_0) -10log((-(ω/ω_0)^2 +1)^2 +(ω/Qω_0)^2)$

Pour tracer le diagramme de Bode qui représente le Gain on doit étudier les 3 cas suivants :

#### `a- w>>ω0`

Le 1 devinet négligeable, on prend $(ω/ω_0)^2$ comme facteur en commun on obtient :

$Gain = 40log(ω/ω_0) -10log((ω/ω_0)^2[(ω/ω_0)^2  +(1/Q))]$

1/Q devinet négligable,donc:

$Gain = 40log(ω/ω_0) -10log((ω/ω_0)^4)$

$Gain = 40log(ω/ω_0) -40log(ω/ω_0) $

>$Gain =0 db$ 

#### `b- w=ω0`

$Gain = 40log(1) -10log((-1 +1)^2 +(1/Q^2))$

$Gain = -10log(1) + 10log(Q^2)$

>$Gain = 20log(Q) = -6 db$

#### `c- w<<ω0`

$(ω/Qω_0)^2$ et $-(ω/ω_0)^2$ deviennent négligeable face au 1, donc :

$Gain = 40log(ω/ω_0) -10log(1)

>$Gain = 40log(ω/ω_0)$
